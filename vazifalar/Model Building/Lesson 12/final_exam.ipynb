# Filling Missing Value

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#funksiya + filling value
def toldir(df):
    for col in df.columns:
        if df[col].dtype=='object':
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:
            df[col].fillna(df[col].mean(),inplace=True)
    return df            
```

```python
toldir(df)
```

```python
df.isnull().sum()
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
encoder=LabelEncoder()
```

```python
def raqamla(df):
    for col in df.columns:
        if df[col].dtype=='object':
           if df[col].nunique() <=5:
                dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
        else:
                df[col]=encoder.fit_transform(df[col])    
    return df            


```

```python
df=raqamla(df)
```

# Model Training Process

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('Heart_Attack_Risk_Moderate',axis=1)
```

```python
y=df['Heart_Attack_Risk_Moderate']
```

```python
#datasetni tarin va test qismarga taqsimlash
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Model Selection

```python
from sklearn.svm import SVC,SVR
```

```python
svc=SVC(kernel='linear',C=1.0)    # boshqa algoritmlardan farqli bunda qavs ichiga parametr beramiz
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

 # Model Evaulation

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
report=classification_report(y_test,y_pred)
```

```python
# Demak son berilgan agar 1 xonali bolsa ozi ,2yo uch honali bolsa raqamlar yigindisi,boshqa hollarda 0 qaytaradigan funksiya tuzish kk

def ifoda_tekshir(raqam):
    raqam=abs(raqam)
    if raqam==1:
        return 1
    if raqam==2 or raqam==3:
        return sum(raqam)
    else:
        return 0
```
