# Encoding

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
encoder=LabelEncoder()
```

```python
cardinality=df.nunique()
```

```python
cardinality            #onehotencoder dan foydalanish faqat 3 ta column un qolgani LabelEncoder
```

```python
def raqamla(df):
    for col in df.columns:
        if df[col].dtype=='object':
            if df[col].nunique()<=5:
                dummies=pd.get_dummies(df[col],prefix=col, dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df   
```

```python
df=raqamla(df)
```

```python
df.info()
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
num_col=df.select_dtypes(include=['int64','float64']).columns.drop('Placement_Yes')
```

```python
num_col
```

```python
def taqribla(df):
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
taqribla (df)
```

# Model Training Process

```python
# Inputlarni x va outputni y bn belgilab ajratib olamiz
x=df.drop('Placement_Yes',axis=1)
y=df['Placement_Yes']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# datalarni qismlarga ajratish
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Model Selection

```python
from sklearn.ensemble import RandomForestClassifier             
```

```python
rf=RandomForestClassifier()
```

```python
rf
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
y_pred
```

```python
x_test
```

# Model Evaluation

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
report=classification_report(y_test,y_pred)
score=accuracy_score(y_test,y_pred)
```

```python
print(report)
print(score)
```
