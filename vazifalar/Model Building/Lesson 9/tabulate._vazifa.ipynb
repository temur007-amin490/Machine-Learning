# Handle missing value with for Loop

```python
def cleaning(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df
```

```python
cleaning(df)
```

```python
df.isnull().sum()
```

# Encoding+ Function

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
def nomerla(df):
    for col in df.columns:
        if df[col].dtype=='object':
            if df[col].nunique()<=5:
                dummies=pd.get_dummies(df[col],prefix=col, dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df           
```

```python
df=nomerla(df)
```

# Model Training Process

```python
# train va testga ajratish kutubxonasini chaqiramiz
from sklearn.model_selection import train_test_split
```

# Inputlarni x va outputni y bn belgilab ajratib olamiz

```python
x=df.drop('Customer_Type_Returning',axis=1)
```

```python
y=df['Customer_Type_Returning']
```

```python
# Datani qismlarga ajratish
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Model Selection

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
```

```python
print(x_train.isnull().sum())     # Input X contains NaN tekshirdik tushib qolgan qiymat yoq
```

```python
x_train = x_train.fillna(x_train.median())
```

```python
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
report=classification_report(y_test,y_pred)
classification_report
```

```python
score=accuracy_score(y_test,y_pred)
score
```
