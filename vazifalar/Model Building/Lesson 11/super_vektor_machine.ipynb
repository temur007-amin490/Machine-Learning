# Filling Missing Value

```python
df.isnull().sum()
```

```python
#hanle missing value + function
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df
```

```python
toldir(df)
```

```python
df.info()
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
encoder=LabelEncoder()
```

```python
def raqamla(df):
    for col in df.columns:
        if df[col].dtype=='object':
            if df[col].nunique()<=5:
                dummies=pd.get_dummies(df[col],prefix=col, dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=raqamla(df)
```

```python
df.info()
```

```python
df.head()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
scaler=StandardScaler()
```

```python
def yaxlitla(df):
    num_col=df.select_dtypes(include=['int64','float64']).columns.drop('Obesity')
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
yaxlitla(df)
```

```python
print(df.columns)
```

# Model Training Process

```python
from sklearn.model_selection import train_test_split
```

```python
# Inputlarni x va outputni y bn belgilab ajratib olamiz
x=df.drop('Obesity',axis=1)
y=df['Obesity']
```

```python
#datasetni tarin va test qismarga taqsimlash
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Model Selection

```python
from sklearn.svm import SVC,SVR
```

```python
svc=SVC(kernel='linear',C=1.0)
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred
```

# Model Evaluation

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
report=classification_report(y_test,y_pred)
```

```python
print(report)  # wauw 96% accuracy score zor natija
```
